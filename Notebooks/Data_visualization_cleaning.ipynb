```python
# Load CSV with correct encoding
df = pd.read_csv(r"D:\Superstore_dataanalysis\Data\Sample - Superstore.csv", encoding='latin1')
```

```python
# Check for duplicates
df.duplicated().sum()
```

```python
# Convert date columns to datetime
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])
```

```python
# Delivery time in days
df['Delivery_Days'] = (df['Ship Date'] - df['Order Date']).dt.days
```

```python
# Profit margin
df['Profit_Margin'] = df['Profit'] / df['Sales']
```

```python
# Year & Month columns for time-based analysis
df['Order_Year'] = df['Order Date'].dt.year
df['Order_Month'] = df['Order Date'].dt.month
```

```python
df.to_csv("D:\Superstore_dataanalysis\Data\Superstore_Cleaned.csv", index=False)
```
